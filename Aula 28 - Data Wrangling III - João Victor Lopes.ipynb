```python
# Ler e combinar todos os arquivos Excel em um único DataFrame
# onde abre e fecha um colchete [] está presente uma list comprehension
# list comprehension = é como se fosse um looping dentro de uma linha só
# Definindo a lista de arquivos Excel
excel_files = ['C:/Users/joaov/Desktop/1 EXCEL - Address parte 1.xlsx', 'C:/Users/joaov/Desktop/1 EXCEL - Address parte 2.xlsx', 'C:/Users/joaov/Desktop/1 EXCEL - Address parte 3.xlsx']

# Ler e combinar todos os arquivos Excel em um único DataFrame
df_combined = pd.concat([pd.read_excel(file) for file in excel_files], ignore_index=True)

# Exibir o DataFrame combinado
df_combined.head()
```

```python
colunas_concatenar = ['AddressLine1, part 1', 'AddressLine1, part 2',
'AddressLine1, part 3', 'AddressLine1, part 4', 'AddressLine1, part 5',
'AddressLine1, part 6', 'AddressLine1, part 7']
```

```python
df_combined[colunas_concatenar] = df_combined[colunas_concatenar].fillna("")
df_combined.head()
```

```python
df_combined['address'] = df_combined[colunas_concatenar].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
df_combined.head()
```

```python
df_combined.drop(columns=colunas_concatenar, inplace=True)
df_combined.head()
```

```python
df_combined['City'] = df_combined['City'].str.replace('õ', 'o').str.replace('é', 'e')
df_combined.head()
```

```python
df_combined['CountryRegion'] = df_combined['CountryRegion'].str.replace('.', ' ', regex=False)
df_combined.head()
```

```python
df_combined['PostalCode'] = df_combined['PostalCode'].str.replace('_', '', regex=False)
df_combined.head()
```

```python
df_combined.info()
```

```python
df_combined['ModifiedDate'] = pd.to_datetime(df_combined['ModifiedDate'])
```

```python
df_combined['AddressLine2'] = df_combined['AddressLine2'].astype(str)
```

```python
df_combined.set_index('AddressID', inplace=True)
```
